```python
import pandas as pd

"""Pertanyaan 1"""
karyawan = {
    'umur': [19,20,31,14,40,44,30,5],
    'gaji': [1900000,2000000,3100000,1400000,4000000,4400000,3000000,500000]
}

df_karyawan = pd.DataFrame(karyawan)

df_karyawan['status'] = df_karyawan['umur'].apply(lambda umur : 'Tua' if umur >= 30 else 'Muda')

"""Pertanyaan 2"""
df_karyawan['gaji_dengan_bonus'] = df_karyawan.apply(lambda df: df['gaji']*1.1 if df['status']=='Tua' else df['gaji']*1.05, axis=1)

"""Pertanyaan 3"""
df_karyawan.info()

"""Pertanyaan 4"""
df_karyawan['gaji_ganda'] = df_karyawan.apply(lambda df: df['gaji']*2 if df['status']=='Tua' else df['gaji'], axis=1)
df_karyawan['gaji_ganda_dengan_bonus'] = df_karyawan.apply(lambda df: df['gaji_ganda']*1.1 if df['status']=='Tua' else df['gaji_ganda']*1.05, axis=1)
df_karyawan
```
