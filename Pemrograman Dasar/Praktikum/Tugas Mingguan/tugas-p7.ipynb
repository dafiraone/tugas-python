```python
import pandas as pd

"""Pengolahan Data"""
df = pd.read_csv('Dataset/TugaspraktikumP7.csv')
df = df.dropna()
df = df.drop_duplicates()
df['produksi_padi'] = df['produksi_padi'].str.replace(',','')
df['produksi_padi'] = pd.to_numeric(df['produksi_padi'])

"""Analisis Statistik"""
total_produksi = df.groupby('nama_kelurahan')['produksi_padi'].sum().reset_index()
total_produksi = total_produksi.sort_values('produksi_padi', ascending=False)
tertinggi = total_produksi.head(5)
terendah = total_produksi.tail(5)

"""Visualisasi Data"""
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2, figsize=(10,5))

ax1.set_title('5 kelurahan dengan total produksi padi tertinggi')
ax1.bar(tertinggi['nama_kelurahan'], tertinggi['produksi_padi'])

ax2.set_title('5 kelurahan dengan total produksi padi terendah')
ax2.bar(terendah['nama_kelurahan'], terendah['produksi_padi'])

plt.tight_layout()
plt.show()
```
