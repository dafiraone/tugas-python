```python
df = pd.read_csv('../Dataset/Latihan_soal.csv')
df = df.dropna()
age_median = df['Age'].median()
score_median = df['Score'].median()

print(age_median)
print(score_median)

import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(22, 4))

ax1.set_title('Age')
ax1.plot(df['Name'], df['Age'])
ax1.axhline(age_median, color='red')

ax2.set_title('Score')
ax2.plot(df['Name'], df['Score'])
ax2.axhline(score_median, color='red')

plt.tight_layout()
plt.show()
```
